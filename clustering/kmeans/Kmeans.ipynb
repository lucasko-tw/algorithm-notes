K-means Example by IPython
---------------------------
此範例為使用 sklearn進行kmeans的運算，並記錄各個參數的意義

```python
# enable online plot in jupyter
%pylab inline

from sklearn.cluster import  KMeans
from sklearn import cluster, datasets
import matplotlib.pyplot as plt  

# load dataset 
iris = datasets.load_iris() 
data = iris.data[: , :2 ]  # we only take the first two features for 2-D .
labels = iris.target 

# Basic K-means Clustering, k=3 
k_means = cluster.KMeans(n_clusters=3)
k_means.fit(data) 
```

### To show the clusters of kmeans
畫出kmeans後，群集情況（使用Kmeans的label標記顏色）

```python
#Plotting predicted labels as a scatter
pl.figure('K-means with 3 clusters, with ')
pl.scatter(data[:, 0], data[:, 1], c=k_means.labels_)
pl.show()
```

### To show the clusters of dataset
使用dataset的label標記顏色
（由此可知，原始資料的分佈狀況，已經很混雜）

```python
#Plotting predicted labels as a scatter
pl.figure('K-means with 3 clusters, compare to original labels')
pl.scatter(data[:, 0], data[:, 1], c=labels)
pl.show()
```

參數設定
------------------------
#### n_clusters：
1. int , 預設值為8
2. 分群的群數，也代表會有多少個centroids

#### max_iter：
1. int, 預設值為300
2. 疊代的次數的最大上限。（疊代次數：當k群各自重新計算一次中心點時，為疊代一次）

#### n_init：
1. int, 預設值為10
2. K-means 起始時會初始化質心（隨機產生），為了避免壞結果，透過運算n_init次，取其中最好的。

#### init：
1. {‘k-means++’, ‘random’ or an ndarray} ， 一種初始化質心的算法，預設值為k-means++
2. 初始化质心（function或array）
3. k-means++：使用特定的方法找尋初始點
4. random：隨機取 k 個資料，作為初始點

#### precompute_distances : 
1. {‘auto’, True, False}，評估是否要將運算載入記憶體中
1. True：若是有足夠的記憶體，並且想要提高計算速度，可將precomputer_distance設為True

#### tol :
1. 最小容忍誤差值，預設值為1e-4，

#### n_jobs：
1. 用於設定進程的數量
2. -1：全部CPUs將被使用，1：將不會使用平行化運算，-2：all CPUs but one are used.

#### random_state：
1. 隨機生成的seed 

#### verbose :
1. int，預設值為0

#### copy_x：
1. boolean，預設為 True
2. copy_x 為 True：原始資料將不會被修改
3. copy_x 為 False：原始資料將會被修改，針對差異值，補上data mean



#### 範例：   


```python
KMeans(copy_x=True, init='k-means++', max_iter=300, n_clusters=3, n_init=10, n_jobs=1, precompute_distances='auto', random_state=None, tol=0.0001, verbose=0)
```
