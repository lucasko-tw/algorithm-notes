### K-means Example by IPython

There is a example for kmeans.


```python
# enable online plot in jupyter
%pylab inline

from sklearn.cluster import  KMeans
from sklearn import cluster, datasets
import matplotlib.pyplot as plt  

# load dataset 
iris = datasets.load_iris() 
data = iris.data[: , :2 ]  # we only take the first two features for 2-D .
labels = iris.target 
```

```python

# Basic K-means Clustering, k=3 
k_means = cluster.KMeans(n_clusters=3)
k_means.fit(data) 
```

```python

#Plotting predicted labels as a scatter
pl.figure('K-means with 3 clusters, with ')
pl.scatter(data[:, 0], data[:, 1], c=k_means.labels_)
pl.show()
```

```python

#Plotting predicted labels as a scatter
pl.figure('K-means with 3 clusters, compare to original labels')
pl.scatter(data[:, 0], data[:, 1], c=labels)
pl.show()
```
